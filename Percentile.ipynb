```python
for i in Descriptive.columns :
    for j in Descriptive.index :
        if(j=="Mean"):
            Descriptive[i][j] = dataset[i].mean()
        elif(j=="Median"):
            Descriptive[i][j] = dataset[i].median()
        else:
            Descriptive[i][j] = dataset[i].mode()[0] 

Descriptive
```

# Percentile

```python
import numpy as np
np.percentile(dataset["ssc_p"],25)
```

```python
#Colums - first,Rows - second 
Descriptive = pd.DataFrame(index = ["Mean", "Median", "Mode","Q1:25(percentile)","Q1:25(describe)","Q2:50","Q3:75","99%","Q4:100"],columns = Quan)
for i in Descriptive.columns :
     Descriptive[i]["Mean"] = dataset[i].mean()
     Descriptive[i]["Median"] = dataset[i].median()
     Descriptive[i]["Mode"] = dataset[i].mode()[0] 
     Descriptive[i]["Q1:25(percentile)"] = np.percentile(dataset[i],25)
     Descriptive[i]["Q1:25(describe)"] = dataset.describe()[i]["25%"]
     Descriptive[i]["Q2:50"] = dataset.describe()[i]["50%"]
     Descriptive[i]["Q3:75"] = dataset.describe()[i]["75%"]
     Descriptive[i]["99%"] = np.percentile(dataset[i],99)
     Descriptive[i]["Q4:100"] = dataset.describe()[i]["max"]
Descriptive
```
